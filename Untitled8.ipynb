```python
numeric_cols = ['math score', 'reading score', 'writing score']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"Outliers in {col}:")
    print(outliers[[col]])
    print("-"*30)
```

```python
for col in numeric_cols:
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
numeric_cols = ['math score', 'reading score', 'writing score']

for col in numeric_cols:
    sns.histplot(df[col], kde=True, bins=5)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in numeric_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
corr = df[numeric_cols].corr()
print(corr)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
for col in numeric_cols:
    sns.boxplot(x=df['test preparation course'], y=df[col])
    plt.title(f"{col} by test preparation course")
    plt.show()
```

```python
sns.countplot(x='test preparation course', data=df)
plt.title("Count of writing score per test preparation course")
plt.show()
```

```python
sns.countplot(x= 'gender', hue='test preparation course', data=df)
plt.title("Count of test preparation course per gender")
plt.show()
```
