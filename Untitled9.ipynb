```python
df=pd.read_csv('/content/retail-store-sales-dirty-for-data-cleaning/retail_store_sales.csv')
```

```python
df.head(20)
```

```python
df['Total Spent']=df['Total Spent'].fillna(df['Total Spent'].mean())
```

```python
df.head(20)
```

```python

df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Total Spent'] / df['Quantity'])
df['Quantity'] = df['Quantity'].fillna(df['Total Spent'] / df['Price Per Unit'])

df['Customer ID'] = df['Customer ID'].astype(str).str.replace('CUST_', '', regex=False)

df['Transaction ID'] = df['Transaction ID'].astype(str).str.replace('TXN_', '', regex=False)

df['Customer ID'] = pd.to_numeric(df['Customer ID'], errors='coerce')
df['Transaction ID'] = pd.to_numeric(df['Transaction ID'], errors='coerce')

expected_spent = df['Price Per Unit'] * df['Quantity']

calculated_discount = df['Total Spent'] < expected_spent

df['Discount Applied'] = df['Discount Applied'].fillna(calculated_discount)

df['Item'] = df.groupby('Category')['Item'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))


df.head(10)

```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(sparse_output=False)

encoded_array = encoder.fit_transform(df[['Category']])

encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['Category']))

df = pd.concat([df.drop('Category', axis=1), encoded_df], axis=1)
```
