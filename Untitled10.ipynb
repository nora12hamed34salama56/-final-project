```python
df=df.drop_duplicates()
```

```python


df['bmi'] = df['bmi'].replace(0, np.nan)

df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```

```python
# تحديد الحد الأقصى عند الـ percentile 99
bmi_upper_limit = df['bmi'].quantile(0.99) # غالباً هتكون في حدود 50-53

# تقليص أي قيمة أعلى من الحد ده
df['bmi'] = df['bmi'].clip(upper=bmi_upper_limit)
```

```python
from numpy._core import numeric
numeric_cols =['age','bmi','HbA1c_level','blood_glucose_level']
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"Outliers in {col}:")
    print(outliers[[col]])
    print("-"*30)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['smoking_history_encoding']=le.fit_transform(df['smoking_history'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])
```
